```python
import mlflow
import mlflow.sklearn
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('datasets/50_startups.csv')
df['State']=df['State'].map({'New York':0,'Florida':1, 'California': 2}).astype(int)
```

```python
# Training Data
X = df[["R&D Spend", "Administration", "Marketing Spend", "State"]]
y = df[["Profit"]]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7,random_state=0)
```

```python
# Import Scikit-learn flavor
import mlflow.sklearn
```

```python
mlflow.create_experiment("Sklearn Model")
# Set the experiment to "Sklearn Model"
mlflow.set_experiment("Sklearn Model")
```

```python
# Set Auto logging for Scikit-learn flavor
mlflow.sklearn.autolog()
```

```python
%%time
lr = LinearRegression(n_jobs=10)
lr.fit(X_train, y_train)
```

```python
# Get a prediction from test data
print(lr.predict(X_test.iloc[[12]]))
```

```python
lr.get_params(deep=True)
```

# Model API

```python
# mlflow.sklearn.load_model('Sklearn Model')
mlflow.sklearn.save_model(lr, 'mlruns/models/lr_model/')
```
